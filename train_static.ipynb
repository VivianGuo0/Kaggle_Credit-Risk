```python
column_list = df.columns.tolist()
isnull_sum = []
for i in column_list:
    isnull_sum.append(df[i].isnull().sum())
df_isnull = pd.DataFrame({"column_name": column_list, "isnull_sum": isnull_sum})
pd.set_option('display.max_rows', 200)
print(df_isnull.describe())
```
